9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
nombres = ['Ana','Luis','Pedro','Sofia','Carmen']
altura = [160,180,175,190,170]

barplot = plt.bar(nombres,altura, color='green')

for bar in barplot:
	altura_b = bar.get_height()
	x_b = bar.get_x()
	w_b = bar.get_width()
	plt.text(x_b + w_b/3.5, altura_b, altura_b, va='bottom')

plt.xlabel("Personas")
plt.ylabel("Alturas")
plt.title("Alturas grupo")

plt.show()
```

```python
plt.style.use('seaborn-v0_8-white')
fig, ax = plt.subplots()
maximo=df['Sold_items_A'].max() #QUIERO DESTACAR EL VALOR MAXIMO 7000(%) para destacarlo (lo separo)
elegido=df.loc[df['Sold_items_A']==maximo,'Year'] # CORRESPONDE AL A칌O 2019
#------------------tengo que conseguir una tupla para pasarla como argumento de la funci칩n
lista=[]
for i in range(0,len(df['Sold_items_A'])):  #recorro los valores y al m맞imo le asigno una separaci칩n
    if df['Sold_items_A'][i]==maximo:
        lista.append(0.2) #la separaci칩n
    else:
        lista.append(0)   #pegados entre s칤
tupla=tuple(lista)  # de [0,0,0,0,0.2] => (0,0,0,0,0.2)
#elegido_destacar = tupla  # Con una tupla y slider escoger uno a destacar # todos pegados con 0 , la sepacion 0.1,0.2
ax.pie(df['Sold_items_A'],explode=tupla, labels=df['Year'],autopct='%1.1f%%',shadow=True, startangle=180);
                                                                                 #pocentaje           el 치ngulo(se ve distinto)
```

```python
plt.pie(df['Sold_items_A'], labels=df['Year'], autopct=lambda p: '{:.0f}\n({:.1f}%)'.format(p * sum(df['Sold_items_A']) / 100, p), startangle=90)
plt.axis('equal')  # Para asegurarse de que el gr치fico sea circular
plt.title("Distribuci칩n de Ventas en el Departamento A por A침o")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))
data = [1000, 3500, 4000, 5500, 7000]
ingredients = ['2015', '2016', '2017', '2018', '2019']
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)  # "{:.1f}%\n({:d})"
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), labels=ingredients)
ax.legend(wedges, ingredients,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=15, weight="bold")
ax.set_title("Matplotlib bakery: A pie")
plt.show()
```
