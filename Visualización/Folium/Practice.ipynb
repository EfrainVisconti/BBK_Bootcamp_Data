# Ejercicios con Folium

#### 0. Importar las librerías necesarias

```python
import folium
import pandas as pd
```

#### 1. Importa los datos de las estaciones de bicicletas de Dublin.

```python
datos = pd.read_csv("data/Dublin_Bike_Station.csv")
```

```python
datos.columns
```

#### 2. Crea un mapa localizado sobre la media de las coordenadas de latitud y longitud.

```python
latlon = datos[['Latitude', 'Longitude']]
latlon.mean()
```

```python
mapa = folium.Map(location=list(latlon.mean()), zoom_start=14)
mapa
```

#### 3. Añade una marca sobre las estaciones de bicicletas respetando las sigueintes indicaciones.
* Las marcas tienen que respetar la siguiente relación de colores:
    * vacío --> rojo
    * lleno --> azul
    * otro --> verde
* Deben llevar el icono de bicicletas.
* Al hacer click debe aparecer el nombre de la estación y el número de bicicletas.

```python
for (index, row) in datos.iterrows():
	if row.loc['Bikes_Available'] == 0 :
		folium.Marker(location = [row.loc['Latitude'], row.loc['Longitude']], popup = row.loc['Name'] + ' ' + str(row.loc['Bikes_Available']), icon=folium.Icon(color='red', icon='bicycle', prefix='fa')).add_to(mapa)
	elif row.loc['Bikes_Available'] == max(datos.Bikes_Available) :
		folium.Marker(location = [row.loc['Latitude'], row.loc['Longitude']], popup = row.loc['Name'] + ' ' + str(row.loc['Bikes_Available']), icon=folium.Icon(color='blue', icon='bicycle', prefix='fa')).add_to(mapa)
	else :
		folium.Marker(location = [row.loc['Latitude'], row.loc['Longitude']], popup = row.loc['Name'] + ' ' + str(row.loc['Bikes_Available']), icon=folium.Icon(color='green', icon='bicycle', prefix='fa')).add_to(mapa)
```
