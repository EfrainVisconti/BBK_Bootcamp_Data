# Agregaciones y agrupaciones

```python
import numpy as np
import pandas as pd
```

### Ej 2. Importar el conjunto de datos desde este [enlace](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Ej 3. Asignarlo a una variable llamada users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

users = pd.read_csv(url, sep='|', header=0, names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])
users
```

### Ej 4. Descubre cuál es la edad media por ocupación.

```python
users.groupby('occupation')['age'].mean()
```

### Ej 5. Descubre el ratio de hombres por ocupación y ordénalo de mayor a menor.

```python
users1 = users.set_index('gender').occupation

ratio = users1[users1.index == 'M'].value_counts() / users.occupation.value_counts()

pd.DataFrame(ratio.sort_values(ascending=False))
```

```python
# javi
dfa = users[users.gender == 'M'].groupby(users.occupation).count() / users.groupby(users.occupation).count()
dfa[['gender']].sort_values(by='gender', ascending=False)
```

### Ej 6. Para cada ocupación, calcular las edades mínima y máxima.

```python
users.groupby('occupation')['age'].aggregate(['min', 'max'])
```

### Ej 7. Para cada combinación de ocupación y género, calcule la edad media.

```python
users.groupby(['gender', 'occupation'])['age'].mean()
```

### Ej 8.  Para cada ocupación presente el porcentaje de mujeres y hombres.

```python
# Clase
(users.groupby('occupation')['gender'].value_counts(1)*100).unstack()
```

```python
# Javi
porcent_m = users[users.gender == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
porcent_f = users[users.gender == 'F'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()

porcentajes = pd.DataFrame({'ocupa_m': porcent_m, 'ocupa_f': porcent_f})
porcentajes
```
