```python
import numpy as np
import pandas as pd
```

### Ej 2. Cree el DataFrame con los siguientes valores:

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

### Ej 3. Asígnalo a una variable llamada regiment.


```python
regiment = pd.DataFrame(raw_data)
regiment
```

### Ej 4. ¿Cuál es la puntuación media de preTestScore del regimiento Nighthawks?

```python
regiment.groupby('regiment')['preTestScore'].mean()['Nighthawks']
```

### Ej 5. Presentar estadísticas generales por empresa.

```python
regiment.groupby('company')[['preTestScore','postTestScore']].describe()
```

### Ej 6. ¿Cuál es la media de preTestScore de cada compañía?

```python
regiment.groupby('company')['preTestScore'].mean()
```

### Ej 7. Presentar la media de preTestScores agrupada por regimiento y empresa.

```python
regiment.groupby(['regiment','company'])['preTestScore'].mean()
```

### Ej 8. Presentar las medias de preTestScores agrupados por regimiento y compañía sin indexación jerárquica.

```python
# Incorrecto
regiment.groupby(['regiment','company'])['preTestScore'].mean().unstack()
```

```python
# Correcto
regiment.groupby(['regiment','company'])['preTestScore'].mean().reset_index()
```

### Ej 9. Agrupe todo el marco de datos por regimiento y compañía.

### Ej 10. ¿Cuál es el número de observaciones en cada regimiento y compañía?

### Ej 11. Itere sobre un grupo e imprima el nombre y todos los datos del regimiento. Crea una función llamada agrupa(df, variable). La variable es la columna utilizada para el agrupamiento.

```python
def agrupa(df, variable):
    agrup=df.groupby(variable)

    for x in agrup:
        print(x)
        print('-'*60)
```

```python
agrupa(regiment,'regiment')
```
