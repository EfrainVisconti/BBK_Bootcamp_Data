13. Crea un array booleano con la condición de ser mujer o hombre, usando la columna `sex`, donde **female** es True. Llámalo `array_fe`.

```python
# Clase
array_fe1 = (titanic["sex"]=="female").values
# array_fe2 = np.where(titanic["sex"] == "female", True, False)
array_fe1
```

```python
array_fe = titanic.sex

def	bool_fun(array_fe):
	new_array = array_fe.copy()
	for i in range(len(array_fe)):
		if array_fe[i] == 'female':
			new_array[i] = True
		else:
			new_array[i] = False

	return (new_array)
```

```python
array_fe = bool_fun(array_fe)
array_fe
```
