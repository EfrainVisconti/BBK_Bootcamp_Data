### Ej 3. Nombre del producto de valor mÃ¡ximo.

```python
price = list(chipo.item_price)

i = 0
while i < len(price):
    price[i] = price[i].rstrip()
    price[i] = price[i].replace('$','')
    price[i] = float(price[i])
    i = i + 1

print(max(price))
```

### Ej 7. Â¿CuÃ¡l fue la cantidad pedida del artÃ­culo mÃ¡s caro? 2 maneras

```python
max = chipo.unit_price.max()
chipo.loc[chipo.unit_price == max]
```

```python
#David
df_prod_car = chipo.loc[chipo.unit_price == 11.89, ['item_name', 'quantity']]
prod_car = df_prod_car.item_name.unique()

prod_cant = {}

for producto in prod_car:
    prod_cant[producto] = df_prod_car.loc[df_prod_car.item_name == producto, ['quantity']].sum()
print(prod_cant)
```

```python
new_chipo = chipo.loc[chipo.unit_price == 11.89, ['item_name', 'quantity']]

#chipo.loc[chipo.item_price == '$44.25 ', ['item_name', 'item_price']]

p1 = new_chipo.loc[new_chipo.item_name == 'Steak Salad Bowl', ['quantity']].sum()
p2 = new_chipo.loc[new_chipo.item_name == 'Carnitas Salad Bowl', ['quantity']].sum()
p3 = new_chipo.loc[new_chipo.item_name == 'Barbacoa Salad Bowl', ['quantity']].sum()

print (f'Steak Salad Bowl: {p1} Carnitas Salad Bowl: {p2} Barbacoa Salad Bowl: {p3}')
```
